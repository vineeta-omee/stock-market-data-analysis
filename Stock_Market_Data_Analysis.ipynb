```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = yf.download("AAPL", start="2022-01-01", end="2024-01-01")
stock.head()
```

```python
stock.info()
```

```python
stock['Close'].plot(figsize=(12,5))
plt.title("Apple Stock Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
stock['MA50'] = stock['Close'].rolling(50).mean()

stock[['Close','MA50']].plot(figsize=(12,5))
plt.title("Closing Price vs 50-Day Moving Average")
plt.show()
```

```python
stock['Returns'] = stock['Close'].pct_change()

stock['Returns'].hist(bins=50)
plt.title("Daily Returns Distribution")
plt.show()
```

```python
stocks = yf.download(['AAPL','TSLA','MSFT'], start="2022-01-01")['Close']

stocks.plot(figsize=(12,6))
plt.title("Stock Price Comparison")
plt.show()
```

Insights:
- Apple stock shows long-term upward trend.
- Moving average smooths volatility.
- Tesla shows higher volatility than Microsoft.
